## Spotify API

```python
import keyring
import os

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

service_name = "spotify"

spotify_client_id = keyring.get_password("spotify", "client_id")
spotify_client_secret = keyring.get_password("spotify", "client_secret")

# If no credentials were stored, prompt the user for them
for cred_name, cred_value in {
    "client_id": spotify_client_id,
    "client_secret": spotify_client_secret,
}.items():
    if cred_value is None:
        api_key = input(f"Please enter your {cred_name}: ")
        keyring.set_password(service_name, cred_name, api_key)

# Export as environment variables
os.environ["SPOTIPY_CLIENT_ID"] = spotify_client_id
os.environ["SPOTIPY_CLIENT_SECRET"] = spotify_client_secret

# Create a Spotify client
auth_manager = SpotifyClientCredentials()
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
# Get the top 20 tracks for the artist "weezer"
results = sp.search(q="weezer", limit=20)
for idx, track in enumerate(results["tracks"]["items"]):
    print(idx, track["name"])
```

```python
# Read the top 10k artists from the Spotify dataset
top_artist_data = pd.read_csv("top10k-spotify-artist-metadata.csv")
top_artists = list(top_artist_data["artist"])
```

```python
# Which features to keep for the artists
artist_features = {
    "external_urls",
    "followers",
    "genres",
    # "href",
    # "id",
    # "images",
    "name",
    "popularity",
    "type",
    "uri",
}
```

```python
# Which features to keep for albums
album_features = {
    # "album_group",
    # "album_type",
    "artists",
    # "available_markets",
    "external_urls",
    # "href",
    # "id",
    # "images",
    "name",
    "release_date",
    "release_date_precision",
    "total_tracks",
    "type",
    "uri",
    "tracks",  # added by us
}
```

```python
# Which features to keep for tracks
track_features = {
    "artists",
    # "available_markets",
    "disc_number",
    "duration_ms",
    "explicit",
    "external_urls",
    # "href",
    # "id",
    "is_local",
    "name",
    # "preview_url",
    "track_number",
    "type",
    "uri",
}
```

```python
from tqdm import tqdm

data_artists = {"artists": []}
data_albums = {"albums": []}


# Search for a artists

for artist_name in tqdm(top_artists):
    results = sp.search(
        q=f"artist:{artist_name}",
        type="artist",
        limit=1,  # avoids unecessary requests. should we use only the top result or not?
    )

    # Skip if no results
    if results["artists"]["items"] is None:
        continue

    # Save artist data
    artist = results["artists"]["items"][0]
    artist_id = artist["id"]
    data_artists["artists"].append({k: artist[k] for k in artist_features})
    # print(artist.keys())

    # Collect album data
    albums = []
    for i_album, album in enumerate(
        sp.artist_albums(artist_id, album_type="album")["items"]
    ):
        # print(album.keys())

        tracks = sp.album_tracks(album["id"])["items"]
        tracks = [{k: track[k] for k in track_features} for track in tracks]
        # print(tracks[0].keys())

        album.update({"tracks": tracks})
        albums.append({k: album[k] for k in album_features})

    data_artists["artists"][-1]["albums"] = albums

# Save the dataset
import json

with open("spotify_top10k_artists.json", "w") as f:
    json.dump(data_artists, f, indent=4)
```
