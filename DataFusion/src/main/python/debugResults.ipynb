```python
import pandas as pd

df = pd.read_csv('../../../data/output/debugResultsDatafusion_recordLevel.csv')
df.head()
```

```python
# sort by longest value in RecordsIDS
df['Group size'] = df['RecordIDS'].str.split('+').apply(len)
```

```python
df = df.sort_values(by=['Group size'], ascending=False)
df.head()
```

```python

for i in range(10):
    df_group = pd.DataFrame()
    row = df.iloc[i]

    df_group['ID'] = row['RecordIDS'].split('+')
    wdc_count = df_group['ID'].str.contains('WDC').value_counts()
    print("WDC\n", wdc_count)

    display(df_group)
```
